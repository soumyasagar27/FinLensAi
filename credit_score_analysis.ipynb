```python
import pandas as pd

# Load data from existing CSV
df = pd.read_csv("data/loan_data.csv.csv")
```

```python
df['MultiLoanFlag'] = df.duplicated(subset='Name', keep=False).astype(int)
df['IncomeGapFlag'] = (df['EMI'] > df['Income']).astype(int)
```

```python
def calculate_credit_score(row):
    score = 100
    if row['LoanAmount'] > row['Income'] * 0.5:
        score -= 20
    if row['MultiLoanFlag'] == 1:
        score -= 25
    if row['IncomeGapFlag'] == 1:
        score -= 30
    return max(score, 0)

df['CreditScore'] = df.apply(calculate_credit_score, axis=1)

def get_risk_category(score):
    if score >= 80:
        return 'Low Risk'
    elif score >= 50:
        return 'Medium Risk'
    else:
        return 'High Risk'

df['RiskCategory'] = df['CreditScore'].apply(get_risk_category)
```
